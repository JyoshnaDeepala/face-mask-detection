```python
import numpy as np
import cv2
```

```python
with_mask = np.load('with.npy')
without_mask = np.load('without.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
#make it 2D from 4D
with_mask = with_mask.reshape(200, 50 * 50 * 3)
without_mask = without_mask.reshape(200, 50 * 50 *3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
x = np.r_[with_mask,without_mask]
```

```python
x.shape
```

```python
labels = np.zeros(x.shape[0])
```

```python
labels[200:] = 1.0
```

```python
names={0:"Mask",1:"No Mask"}
```

```python
pip install scikit-learn
```

```python
from sklearn.svm import SVC
#svc- support vector classifier
from sklearn.metrics import accuracy_score
# to get the accuracy of trained model
```

```python
#to test and train divide using this package
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,labels, test_size=0.25)
```

```python
x_train.shape
```

```python
#so many diamentions there , so make it smooth reduce them
#the math behind this eigan value/eigan vector
from sklearn.decomposition import PCA
#PCA- principal component anaysis
```

```python
pca = PCA(n_components=3)
x_train = pca.fit_transform(x_train)
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
svm = SVC()
svm.fit(x_train,y_train)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
